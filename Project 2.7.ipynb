```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plot
```

```python
dataset= pd.read_csv ('https://raw.githubusercontent.com/CunyLaguardiaDataAnalytics/datasets/master/2014-15_To_2016-17_School-_Level_NYC_Regents_Report_For_All_Variables.csv')
```

```python
#Columns within the Dataset
dataset.head(2)
```

```python
#Question - What is the number of students tested by year in each 'School level' ?
#Answer - 'HighSchool' has the Highest testing count year over year, Followed by 'Secondary School'
Table1=dataset.groupby(['Year','School Level'])['Total Tested'].sum().sort_values(ascending=False)
bar1=dataset.groupby(['Year','School Level'])['Total Tested'].sum().sort_values(ascending=True).plot.barh(figsize=(6, 6))
print(Table1)
print(bar1)
```

```python
#Question - What was the Year in which the highest overall testing happended ?
#Answer - 2015 was the year in which the most 
# Chart and table to Display the Year over Year count of the tested students. 
Table2=dataset.groupby(['Year'])['Total Tested'].sum().sort_values(ascending=False)
pie=dataset.groupby(['Year'])['Total Tested'].sum().sort_values(ascending=False).plot.pie(figsize=(6, 6))
print(Table2)
print(pie)
```

```python
#Question - What were the TOP 3 exams categories which received the highest test?
#Answer - 'Common Core Algebra', 'G'lobal History and Geography' & 'Living Environment' where the TOP 3 exam categories.
Table3=dataset.groupby(['Regents Exam'])['Total Tested'].sum().sort_values(ascending=False)
Bar3=dataset.groupby(['Regents Exam'])['Total Tested'].sum().sort_values(ascending=False).plot.bar(figsize=(10, 5))
print(Table3)
print(Bar3)
```

```python
#Question - Which School has the Highest Test type count.
#Answer - 'Brooklyn Technical High School' had the highest count of test recorded. 

Table4=dataset.groupby(['School Name','Regents Exam'])['Total Tested'].sum().sort_values(ascending=False)
print(Table4)
```
